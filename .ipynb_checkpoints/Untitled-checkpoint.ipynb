```python
# data preprocessing and cleaning

import os
import cv2
import numpy as np
from tqdm import tqdm
from PIL import Image

# Paths
data_dir = "C:/Users/Vidhi/Downloads/dataset"
categories = ["beginner", "intermediate", "advanced"]

# Image size for resizing
img_size = 224

# To store image data and labels
data = []

for category in categories:
    category_path = os.path.join(data_dir, category)
    for asana in os.listdir(category_path):
        asana_path = os.path.join(category_path, asana)
        for img_name in tqdm(os.listdir(asana_path), desc=f"Processing {asana}"):
            try:
                img_path = os.path.join(asana_path, img_name)
                img = Image.open(img_path).convert("RGB")
                img = img.resize((img_size, img_size))
                img_array = np.array(img) / 255.0  # Normalize
                label = asana
                data.append([img_array, label])
            except Exception as e:
                print(f"Skipped {img_path} due to {e}")

# Shuffle the data
import random
random.shuffle(data)

# Splitting features and labels
X = []
y = []

for features, label in data:
    X.append(features)
    y.append(label)

X = np.array(X)
y = np.array(y)
```

```python
# data preprocessing and cleaning

import os
import shutil
import random
from PIL import Image
from pathlib import Path
from tqdm import tqdm

# ----------------------
# üìÅ Step 1: Set Paths
# ----------------------

main_dataset_path = r"C:\Users\Vidhi\Downloads\dataset"  # Your original dataset
output_base_path = r"C:\Users\Vidhi\Downloads\dataset_split"  # Where to save the split folders
img_size = (224, 224)  # Resize target

# ----------------------
# üí° Step 2: Split Ratio
# ----------------------
split_ratio = {
    "train": 0.7,
    "val": 0.2,
    "test": 0.1
}

# ----------------------
# üîÑ Step 3: Clean & Process Images
# ----------------------
def clean_and_resize_image(src_path, dst_path):
    try:
        with Image.open(src_path) as img:
            img = img.convert("RGB")
            img = img.resize(img_size)
            img.save(dst_path)
    except Exception as e:
        print(f"‚ùå Failed on {src_path}: {e}")

# ----------------------
# üìÇ Step 4: Process Each Subfolder (Beginner/Intermediate/Advanced)
# ----------------------
if not os.path.exists(output_base_path):
    os.makedirs(output_base_path)

levels = os.listdir(main_dataset_path)

for level in levels:
    level_path = os.path.join(main_dataset_path, level)
    
    if not os.path.isdir(level_path):
        continue

    asana_folders = os.listdir(level_path)

    for asana in tqdm(asana_folders, desc=f"Processing {level}"):
        asana_path = os.path.join(level_path, asana)
        all_images = [f for f in os.listdir(asana_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        random.shuffle(all_images)

        train_cut = int(len(all_images) * split_ratio["train"])
        val_cut = train_cut + int(len(all_images) * split_ratio["val"])

        split_data = {
            "train": all_images[:train_cut],
            "val": all_images[train_cut:val_cut],
            "test": all_images[val_cut:]
        }

        for split in ["train", "val", "test"]:
            split_folder = os.path.join(output_base_path, split, level, asana)
            os.makedirs(split_folder, exist_ok=True)

            for img_name in split_data[split]:
                src_img_path = os.path.join(asana_path, img_name)
                dst_img_path = os.path.join(split_folder, img_name)
                clean_and_resize_image(src_img_path, dst_img_path)

print("\n‚ú® Data preprocessing and splitting complete!")
```

```python
import os
import shutil
import random
from tqdm import tqdm

def split_dataset(dataset_dir='dataset', output_dir='dataset', train_split=0.7, val_split=0.15):
    levels = ['Beginner', 'Intermediate', 'Advanced']

    for level in levels:
        level_dir = os.path.join(dataset_dir, level)
        if not os.path.exists(level_dir):
            print(f"Level not found: {level_dir}")
            continue
        
        asanas = os.listdir(level_dir)
        for asana in tqdm(asanas, desc=f"Processing {level}"):
            asana_path = os.path.join(level_dir, asana)
            if not os.path.isdir(asana_path):
                continue
            
            images = os.listdir(asana_path)
            random.shuffle(images)
            
            total = len(images)
            train_count = int(total * train_split)
            val_count = int(total * val_split)
            test_count = total - train_count - val_count
            
            splits = {
                'train': images[:train_count],
                'val': images[train_count:train_count+val_count],
                'test': images[train_count+val_count:]
            }

            for split, files in splits.items():
                dest_dir = os.path.join(output_dir, split, asana)
                os.makedirs(dest_dir, exist_ok=True)
                for file in files:
                    src_path = os.path.join(asana_path, file)
                    dst_path = os.path.join(dest_dir, file)
                    shutil.copy2(src_path, dst_path)
                    
    print("‚ú® Data preprocessing and splitting complete!")

# Run it
split_dataset(dataset_dir='dataset')
```

```python
# Model Building

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import os

# Set parameters
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 15
BASE_DIR = r"C:\Users\Vidhi\Downloads\dataset_split"  # Path to your dataset folder

# 1. Image Augmentation + Load Data
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    os.path.join(BASE_DIR, 'train'),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_data = val_datagen.flow_from_directory(
    os.path.join(BASE_DIR, 'val'),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

# 2. Base Model: MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base_model.trainable = False

# 3. Custom Top Layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(train_data.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# 4. Compile
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 5. Train
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS
)

# 6. Save Model
model.save("yoga_pose_model.h5")

# 7. Plot Accuracy and Loss
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.show()

from tensorflow.keras.models import load_model

model_path = "yoga_pose_model.h5"  # ‚úÖ Only if it's in current folder
model = load_model(model_path)
```

```python
import cv2
import numpy as np
import json
import time
import tkinter as tk
from tkinter import messagebox
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import Image, ImageTk

# üíæ Load the trained model
yoga_model = load_model("yoga_pose_model.h5")

# üìú Load the asana details JSON
#data_path = r"‚Ä™C:/Users/Vidhi/Downloads/dataset/yogi.json"  # Make sure this is in the same directory or give full path
with open(r"C:/Users/Vidhi/Downloads/dataset/yogi.json", "r", encoding="utf-8") as file:
    asana_data = json.load(file)

# üè∑Ô∏è Get all possible classes (levels)
levels = ["Beginner", "Intermediate", "Advanced"]

# üéØ Predict level using the model
def predict_level(frame):
    resized = cv2.resize(frame, (224, 224))
    normalized = resized / 255.0
    img_array = np.expand_dims(normalized, axis=0)
    pred = yoga_model.predict(img_array)
    class_index = np.argmax(pred)
    confidence = pred[0][class_index] * 100
    return levels[class_index], confidence

# üîç Get matching asana from JSON (based on predicted level and matching name in folder if needed)
def get_asana_by_level(level):
    matching_asanas = [a for a in asana_data if a["level"] == level]
    if matching_asanas:
        return np.random.choice(matching_asanas)
    return None

# üé® Tkinter UI setup
root = tk.Tk()
root.title("üßò‚Äç‚ôÄÔ∏è YogiSense Pose Detector")
root.configure(bg="#ffe6f0")  # Pastel pink

# üñºÔ∏è Frame to hold webcam
frame_holder = tk.Label(root)
frame_holder.pack()

# üìç Status label
status_label = tk.Label(root, text="Get into your pose...", font=("Helvetica", 14), fg="#1f2f60", bg="#ffe6f0")
status_label.pack(pady=10)

# üé• Start video capture
cap = cv2.VideoCapture(0)

start_time = time.time()
duration = 30  # seconds to keep webcam running

# ü™Ñ Function to update frame in UI
def update_frame():
    ret, frame = cap.read()
    if not ret:
        status_label.config(text="Camera not detected")
        return

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img = Image.fromarray(frame_rgb)
    imgtk = ImageTk.PhotoImage(image=img)
    frame_holder.imgtk = imgtk
    frame_holder.configure(image=imgtk)

    if time.time() - start_time < duration:
        root.after(100, update_frame)
    else:
        cap.release()
        cv2.destroyAllWindows()
        analyze_pose(frame_rgb)

# üìä Analyze the final pose and display popup
def analyze_pose(final_frame):
    predicted_level, confidence = predict_level(final_frame)
    asana = get_asana_by_level(predicted_level)

    if asana:
        msg = f"‚ú® Asana: {asana['english_name']} ({asana['sanskrit_name']})\n"
        msg += f"üìà Level: {predicted_level} ({confidence:.2f}% confidence)\n\n"
        msg += "‚úÖ Benefits:\n" + "\n".join(f"‚Ä¢ {b}" for b in asana['benefits']) + "\n\n"
        msg += "‚ö†Ô∏è Cautions:\n" + "\n".join(f"‚Ä¢ {c}" for c in asana['cautions']) + "\n\n"
        msg += f"‚ñ∂Ô∏è Reference Video:\n{asana['youtube']}"

        messagebox.showinfo("üéâ Perfect Pose!", msg)
        status_label.config(text="Great job! You did it! üßò")
    else:
        messagebox.showwarning("Oops!", "Couldn't find matching asana data.")
        status_label.config(text="Try again with a clear pose.")

# ‚ñ∂Ô∏è Start the detection
update_frame()
root.mainloop()
```
